```python
import opendatasets as od 
import pandas as pd

od.download("https://www.kaggle.com/datasets/ghousethanedar/smiledetection")
```

```python
#data preprocessing
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.image import img_to_array
import keras.utils as util
import imutils
from imutils import paths
from tensorflow.keras.preprocessing.image import load_img
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import os

tf.random.set_seed(0)
np.random.seed(0)
DIRECTORY = r"C:\dev\aicode3\smiledetection\datasets"
CATEGORIES = ["not_smiling", "smiling"]

X = [] #data
y = [] #target

for category in CATEGORIES:
    path = os.path.join(DIRECTORY, category)
    for img in os.listdir(path):
        img_path = os.path.join(path,img)
        image = load_img(img_path, color_mode='rgb', target_size=(47, 67))
        image = img_to_array(image)
        X.append(image)
        label = img_path.split(os.path.sep)[-2]
        label = 'smiling' if label == 'smiling' else 'not_smiling'
        y.append(label)


X = np.array(X) 
y = np.array(y)



    
```

```python
from sklearn.model_selection import train_test_split
(X_tn, X_te0, y_tn, y_te0) = train_test_split(X, y, test_size = 0.20, random_state=1)
X_tn = X_tn/255
X_te = X_te0/255
#making them values from 0 to 1
plt.figure(figsize=(15,8))
for i in range(2*4):
    plt.subplot(2,4, i+1)
    plt.imshow(X_tn[i])
    plt.title(y_tn[i])
plt.show()

```

```python
#one-hot encoding
le_tn = LabelEncoder().fit(y_tn)
le_te = LabelEncoder().fit(y_te0)
y_tn = util.to_categorical(le_tn.transform(y_tn), 2) # 0 -> not smiling, 1 -> smiling
y_te = util.to_categorical(le_te.transform(y_te0), 2) 
```

```python
#careful when running...might mess up the percentage 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D
from tensorflow.keras.layers import MaxPool2D, Flatten
from tensorflow.keras.layers import Dropout

n = 2

model = Sequential()
model.add(Conv2D(32, kernel_size=(5,5), input_shape=(47,67,3), padding='valid', activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Dropout(0.25))
model.add(Conv2D(32, kernel_size=(3,3), input_shape=(47,67,3), padding ='valid',  activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(n, activation ='softmax'))
model.summary()
```

```python
from tensorflow.keras.models import load_model
# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# #training
# history = model.fit(X_tn, y_tn, epochs=2, batch_size=188) 
# model.save('smile_detector_model.keras')
#above code was used for training the model and the code below uses the trained model directly
cnn_model = load_model('smile_detector_model.keras')
cnn_model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
history= cnn_model.fit(X_tn, y_tn, epochs=2, batch_size=188)
```
